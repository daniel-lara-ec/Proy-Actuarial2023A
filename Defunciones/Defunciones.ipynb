```python
# list all sav files in the directory
archivos = glob.glob('*.sav')
# get year from file name
years = [int(x.split('_')[1][:-4]) for x in archivos]
```

```python
# create a ditcitonatry with archivos and years
diccionario_archivos = dict(zip(archivos, years))
```

```python
bloques = []

for archivo in archivos:
    bloques.append(pd.read_spss(archivo))
```

```python
bloques[0].head()
```

```python
bloques[0]['mesf'].unique()
```

```python
tabla = bloques[0]

contador = 0

#delete first thee characters of column provf
tabla['edad'] = tabla['edad'].astype(int)
tabla['provf'] = tabla['provf'].str[:2]
tabla['mesf'] = tabla['mesf'].astype(int)
tabla['provf'] = tabla['provf'].astype(int)
tabla['sexo'] = tabla['sexo'].str[2:]

provincias = tabla['provf'].unique()
sexo = tabla['sexo'].unique()

# create folder for each year

os.mkdir(str(years[contador]))

for provincia in provincias:

    df_provincia = tabla[(tabla['provf'] == provincia) & (tabla['mesf'] == 12)]
    #
    fallecimientos_homres = df_provincia[df_provincia['sexo'] == 'HOMBRES']['edad'].value_counts().to_frame()
    fallecimientos_mujeres = df_provincia[df_provincia['sexo'] == 'MUJERES']['edad'].value_counts().to_frame()
    #
    fallecimientos_homres.sort_index(inplace=True)
    fallecimientos_mujeres.sort_index(inplace=True)

    fallecimientos_homres.rename(columns={'count': 'Hombres'}, inplace=True)
    fallecimientos_mujeres.rename(columns={'count':'Mujeres'}, inplace=True)
    # join dataframes
    fallecimientos = fallecimientos_homres.join(fallecimientos_mujeres, how='outer')

    fallecimientos.fillna(0, inplace=True)
    fallecimientos['Hombres'] = fallecimientos['Hombres'].astype(int)
    fallecimientos['Mujeres'] = fallecimientos['Mujeres'].astype(int)

    fallecimientos['Total'] = fallecimientos['Hombres'] + fallecimientos['Mujeres']

    archivo_nombre = 'fallecimientos_'+str(diccionaro_provincias[provincia][0])+'.xlsx'

    archivo_nombre = archivo_nombre.replace(' ','')

    fallecimientos.to_excel(str(years[contador])+'/'+archivo_nombre)
```

```python
df = pd.read_csv('defunciones2019.csv',sep=',')
```

```python
df.head()
```

```python
# drop Unnamed: 0 column
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# replace 1 with Hombre and 2 with Mujer in column sexo
df['sexo'] = df['sexo'].replace(1,'Hombre')
df['sexo'] = df['sexo'].replace(2,'Mujer')
```

```python
df.head()
```

```python
# rename column prov_res to prov
df.rename(columns={'prov_res':'prov'}, inplace=True)
```

```python
diccionaro_provincias = {1: ['AZUAY'], 2: ['BOLIVAR'], 3: ['CAÑAR'], 4: ['CARCHI'], 5: [' COTOPAXI'], 6: ['CHIMBORAZO'], 7: ['EL ORO'], 8: ['ESMERALDAS'], 9: ['GUAYAS'], 10: ['IMBABURA'], 11: ['LOJA'], 12: ['LOS RIOS'], 13: ['MANABI'], 14: [' MORONA SANTIAGO'], 15: [' NAPO'], 16: [' PASTAZA'], 17: ['PICHINCHA'], 18: ['TUNGURAHUA'], 19: [' ZAMORA CHINCHIPE'], 20: [' GALAPAGOS'], 21: ['SUCUMBIOS'], 22: [' ORELLANA'], 23: ['SANTO DOMINGO DE LOS TSACHILAS'], 24: ['SANTA ELENA'], 90: ['ZONAS NO DELIMITADAS'],88:['EXTERIOR']}
```

```python
# replace prov with name  from diccionario_provincias
df['prov'] = df['prov'].replace(diccionaro_provincias)
```

```python
df.columns
```

```python
# modify df
df['prov'] = df['prov'].replace(' ','')
df['prov'] = df['prov'].replace('Á','A')
df['prov'] = df['prov'].replace('É','E')
df['prov'] = df['prov'].replace('Í','I')
df['prov'] = df['prov'].replace('Ó','O')
df['prov'] = df['prov'].replace('Ú','U')
df['prov'] = df['prov'].replace('á','a')
df['prov'] = df['prov'].replace('é','e')
df['prov'] = df['prov'].replace('í','i')
df['prov'] = df['prov'].replace('ó','o')
df['prov'] = df['prov'].replace('ú','u')
```

```python
provincias = df['prov'].unique()
```

```python
provincias
```

```python
for provincia in provincias:
    
    df_provincia = df[df['prov'] == provincia]
    
    fallecimientos_hombres = df_provincia[df_provincia['sexo'] == 'Hombre']['edad'].value_counts().to_frame()
    fallecimientos_mujeres = df_provincia[df_provincia['sexo'] == 'Mujer']['edad'].value_counts().to_frame()

    fallecimientos_hombres.sort_index(inplace=True)
    fallecimientos_mujeres.sort_index(inplace=True)

    fallecimientos_hombres.rename(columns={'count': 'Hombres'}, inplace=True)
    fallecimientos_mujeres.rename(columns={'count':'Mujeres'}, inplace=True)

    fallecimientos = fallecimientos_hombres.join(fallecimientos_mujeres, how='outer')

    fallecimientos.fillna(0, inplace=True)

    fallecimientos['Hombres'] = fallecimientos['Hombres'].astype(int)
    fallecimientos['Mujeres'] = fallecimientos['Mujeres'].astype(int)

    fallecimientos['Total'] = fallecimientos['Hombres'] + fallecimientos['Mujeres']

    archivo_nombre = 'fallecimientos_'+provincia+'.xlsx'

    archivo_nombre = archivo_nombre.replace(' ','')

    fallecimientos.to_excel('2019/'+archivo_nombre)
    
```
