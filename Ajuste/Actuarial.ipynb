```python
#Suavizamiento
import math
def coef(s):
  coefi=np.zeros(s+1)
  for k in range(s+1):
    coefi[k]=(-1)**(k)*math.factorial(s)/(math.factorial(k)*(math.factorial(s-k)))
  return coefi
```

```python
#Función para la matriz D^T *D según los valores de N y s.
def mat_DTD(N,s):
  D=np.zeros([N-s, N])
  coefi=coef(s)
  for i in range(N-s):
    for j in range(s+1):
      D[i,j+i]=coefi[j]

  F=np.matmul(D.transpose(),D)
  return F
```

```python
#Matriz W:
w_n=np.ones(df2["qxH"].size)
W=np.diag(w_n)
W
```

```python
DTD=mat_DTD(df2["qxH"].size,2)
lamb=1.5
q_suav=np.matmul(np.linalg.inv(W+lamb*DTD),W@df2["qxH"])
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df2["Edad"], y=q_suav,mode='lines',name='Hombres suave'))
fig.add_trace(go.Scatter(x=df2["Edad"], y=df2["qxH"],mode='lines',name='Hombres observado'))
fig.show()
```

```python
DTD=mat_DTD(df2["qxM"].size,2)
lamb=1.5
q_suavM=np.matmul(np.linalg.inv(W+lamb*DTD),W@df2["qxM"])

fig = go.Figure()
fig.add_trace(go.Scatter(x=df2["Edad"], y=q_suavM, mode='lines',name='Mujeres suave'))
fig.add_trace(go.Scatter(x=df2["Edad"], y=df2["qxM"],mode='lines',name='Mujeres observado'))
fig.show()
```

```python
DTD=mat_DTD(df2["qxT"].size,2)
lamb=1.5
q_suavT=np.matmul(np.linalg.inv(W+lamb*DTD),W@df2["qxT"])

fig = go.Figure()
fig.add_trace(go.Scatter(x=df2["Edad"], y=q_suavT, mode='lines',name='Total suave'))
fig.add_trace(go.Scatter(x=df2["Edad"], y=df2["qxT"],mode='lines',name='Total observado'))
fig.show()
```

```python
#Defino las columnas para los qx a suavizar con media geométrica móvil, n=3
df2["qxH_c"]=np.zeros(df2["qxH"].size)
df2["qxM_c"]=np.zeros(df2["qxM"].size)
df2["qxT_c"]=np.zeros(df2["qxT"].size)

#Defino las columnas para los ln(qx) a suavizar con media geométrica móvil, n=3
#Se suavizarán los valores de lnqx pues son más sensibles a cambios abruptos en pequeños periodos
df2["lnqxH"]=-np.log(df2["qxH"])
df2["lnqxM"]=-np.log(df2["qxM"])
df2["lnqxT"]=-np.log(df2["qxT"])
df2["lnqxH_c"]=np.zeros(df2["lnqxH"].size)
df2["lnqxM_c"]=np.zeros(df2["lnqxM"].size)
df2["lnqxT_c"]=np.zeros(df2["lnqxT"].size)
```

```python
#Medias geométricas móviles, n=3 para qx
df2["qxH_c"][0]=stat.geometric_mean(df2["qxH"][0:2])
for i in range(1,df2["qxH_c"].size):
    df2["qxH_c"][i]=stat.geometric_mean(df2["qxH"][i-1:i+2])

df2["qxM_c"][0]=stat.geometric_mean(df2["qxM"][0:2])
for i in range(1,df2["qxM_c"].size):
    df2["qxM_c"][i]=stat.geometric_mean(df2["qxM"][i-1:i+2])
    
df2["qxT_c"][0]=stat.geometric_mean(df2["qxT"][0:2])
for i in range(1,df2["qxT_c"].size):
    df2["qxT_c"][i]=stat.geometric_mean(df2["qxT"][i-1:i+2])
```

```python
#Medias geométricas móviles, n=3 para lnqx
df2["lnqxH_c"][0]=-stat.geometric_mean(df2["lnqxH"][0:2])
for i in range(1,df2["lnqxH_c"].size-2):
    df2["lnqxH_c"][i]=-stat.geometric_mean(df2["lnqxH"][i-1:i+2])
df2["lnqxH_c"][97]=df2["lnqxH_c"][96]
df2["lnqxH_c"][98]=0
    
    
df2["lnqxM_c"][0]=-stat.geometric_mean(df2["lnqxM"][0:2])
for i in range(1,df2["lnqxM_c"].size-2):
    df2["lnqxM_c"][i]=-stat.geometric_mean(df2["lnqxM"][i-1:i+2])
df2["lnqxM_c"][97]=df2["lnqxM_c"][96]
df2["lnqxM_c"][98]=0    

    
df2["lnqxT_c"][0]=-stat.geometric_mean(df2["lnqxT"][0:2])
for i in range(1,df2["lnqxT_c"].size-2):
    df2["lnqxT_c"][i]=-stat.geometric_mean(df2["lnqxT"][i-1:i+2])
df2["lnqxT_c"][97]=df2["lnqxT_c"][96]
df2["lnqxT_c"][98]=0
```
