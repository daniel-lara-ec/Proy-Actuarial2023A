# Carga de datos

```python
df = pd.read_csv('CPV2010M_Poblacion.csv')
```

```python
df_del_poli = pd.read_csv("DiccionarioDelimitacionPolitica.csv",sep=';')
```

```python
df_del_poli.head()
```

```python
df_provincias = df_del_poli.copy()
```

```python
# rome duplicates and keep the first
df_provincias.drop_duplicates(subset='Código Provincia', keep='first', inplace=True)
df_provincias = df_provincias[['Código Provincia','Nombre de la Provincia']]
```

```python
df_provincias.head(25)
```

```python
# df_provincias to dict
dict_provincias = df_provincias.set_index('Código Provincia').T.to_dict('list')
```

```python
print(dict_provincias)
```

```python
provincias_amazonia = [14,15,16,19,21,22]
```

```python
# print provincias name
for provincia in provincias_amazonia:
    print(dict_provincias[provincia][0])
    
```

```python
for provincia in [17]:
    df_provincia = df[df['I01']==provincia]
    # Transform to dataframe 
    edad_provincia_h = df_provincia[df_provincia['P01']==1]['P03'].value_counts().to_frame()
    edad_provincia_m = df_provincia[df_provincia['P01']==2]['P03'].value_counts().to_frame()
    # order index
    edad_provincia_h.sort_index(inplace=True)
    edad_provincia_m.sort_index(inplace=True)

    # rename columns count
    edad_provincia_h.rename(columns={'count':'Hombres'}, inplace=True)
    edad_provincia_m.rename(columns={'count':'Mujeres'}, inplace=True)

    # join dataframes
    edad_provincia = edad_provincia_h.join(edad_provincia_m, how='outer')
    # convert to int 
    edad_provincia = edad_provincia.astype(int)
    # add column total
    edad_provincia['Total'] = edad_provincia['Hombres'] + edad_provincia['Mujeres']

    archivo_nombre = 'provincia_'+str(dict_provincias[provincia][0])+'.xlsx'
    # remove spaces in name
    archivo_nombre = archivo_nombre.replace(' ','')

    edad_provincia.to_excel(archivo_nombre)
```

```python
region_costa_insular_codigo = [7,8,9,12,13,24,20,23]

for provincia in region_costa_insular_codigo:
    df_provincia = df[df['I01']==provincia]
    # Transform to dataframe 
    edad_provincia_h = df_provincia[df_provincia['P01']==1]['P03'].value_counts().to_frame()
    edad_provincia_m = df_provincia[df_provincia['P01']==2]['P03'].value_counts().to_frame()
    # order index
    edad_provincia_h.sort_index(inplace=True)
    edad_provincia_m.sort_index(inplace=True)

    # rename columns count
    edad_provincia_h.rename(columns={'count':'Hombres'}, inplace=True)
    edad_provincia_m.rename(columns={'count':'Mujeres'}, inplace=True)

    # join dataframes
    edad_provincia = edad_provincia_h.join(edad_provincia_m, how='outer')
    # replace NaN with 0
    edad_provincia.fillna(0, inplace=True)
    # convert to int 
    edad_provincia = edad_provincia.astype(int)
    # add column total
    edad_provincia['Total'] = edad_provincia['Hombres'] + edad_provincia['Mujeres']

    archivo_nombre = 'provincia_'+str(provincia)+'.xlsx'
    # remove spaces in name
    archivo_nombre = archivo_nombre.replace(' ','')

    edad_provincia.to_excel(archivo_nombre)
    
```
