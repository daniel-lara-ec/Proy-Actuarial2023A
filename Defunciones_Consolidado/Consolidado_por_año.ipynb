```python
# list all excel files in current directory with glob   
archivos = glob.glob('*.xlsx')
archivos
```

```python
# split the name with _ and get the first part
nombres = [x.split('_')[0] for x in archivos]
nombres
```

```python
# create dictionary with the nombres and numbers from 1 to 12
diccionario = dict(zip(range(14),nombres))
diccionario
```

```python
region_amazonia = ['MoronaSantiago', 'Napo', 'Pastaza', 'ZamoraChinchipe', 'Orellana', 'Sucumbios']
region_costa_insular = ['ElOro', 'Esmeraldas', 'Guayas', 'LosRios', 'Manabi', 'SantaElena','Galapagos','SantoDomingoDeLosTsachilas']
region_amazonia_codigo = [14,15,16,19,22,21]
region_costa_insular_codigo = [7,8,9,12,13,24,20,23]
```

```python
bloque = []

for archivo in archivos:
    bloque.append(pd.read_excel(archivo))
```

```python
lista = list(range(151))
amazonia_hombres = pd.DataFrame(index=lista)
```

```python
# create table with index number from 0 to 150
lista = list(range(151))
amazonia_hombres = pd.DataFrame(index=lista)
amazonia_mujeres = pd.DataFrame(index=lista)
costa_insular_hombres = pd.DataFrame(index=lista)
costa_insular_mujeres = pd.DataFrame(index=lista)

for i in range(14):

    tabla = bloque[i]

    tabla['Hombres'] = tabla['Hombres'].astype(int)
    tabla['Mujeres'] = tabla['Mujeres'].astype(int)

    # set edad as index
    tabla.set_index('edad', inplace=True)


    if int(diccionario[i]) in region_amazonia_codigo:

        amazonia_hombres = amazonia_hombres.join(tabla['Hombres'], how='outer')
        amazonia_mujeres = amazonia_mujeres.join(tabla['Mujeres'], how='outer')

        # fill NaN with 0
        amazonia_hombres.fillna(0, inplace=True)
        amazonia_mujeres.fillna(0, inplace=True)

        # rename column with region name
        amazonia_hombres.rename(columns={'Hombres':'Hombres'+diccionario[i]}, inplace=True)
        amazonia_mujeres.rename(columns={'Mujeres':'Hombres'+diccionario[i]}, inplace=True)

        # change type to int
        amazonia_hombres['Hombres'+diccionario[i]] = amazonia_hombres['Hombres'+diccionario[i]].astype(int)
        amazonia_mujeres['Hombres'+diccionario[i]] = amazonia_mujeres['Hombres'+diccionario[i]].astype(int)
        
    elif int(diccionario[i]) in region_costa_insular_codigo:

        costa_insular_hombres = costa_insular_hombres.join(tabla['Hombres'], how='outer')
        costa_insular_mujeres = costa_insular_mujeres.join(tabla['Mujeres'], how='outer')

        # fill NaN with 0
        costa_insular_hombres.fillna(0, inplace=True)
        costa_insular_mujeres.fillna(0, inplace=True)

        # rename column with region name
        costa_insular_hombres.rename(columns={'Hombres':'Hombres'+diccionario[i]}, inplace=True)
        costa_insular_mujeres.rename(columns={'Mujeres':'Hombres'+diccionario[i]}, inplace=True)

        # change type to int
        costa_insular_hombres['Hombres'+diccionario[i]] = costa_insular_hombres['Hombres'+diccionario[i]].astype(int)
        costa_insular_mujeres['Hombres'+diccionario[i]] = costa_insular_mujeres['Hombres'+diccionario[i]].astype(int)        
```

```python
# compute total for each region
amazonia_hombres['Total'] = amazonia_hombres.sum(axis=1)
amazonia_mujeres['Total'] = amazonia_mujeres.sum(axis=1)
costa_insular_hombres['Total'] = costa_insular_hombres.sum(axis=1)
costa_insular_mujeres['Total'] = costa_insular_mujeres.sum(axis=1)
```

```python
# save file to excel 
amazonia_hombres.to_excel(f'A_{anio}_h.xlsx',index=False)
amazonia_mujeres.to_excel(f'A_{anio}_m.xlsx',index=False)
costa_insular_hombres.to_excel(f'CI_{anio}_h.xlsx',index=False)
costa_insular_mujeres.to_excel(f'CI_{anio}_m.xlsx',index=False)
```
